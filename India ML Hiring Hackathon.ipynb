```python
#Dropping "loan_id since it is just the sequence value"

#train_data.isnull().sum()   ## checking if any entry is null
train_mod=train_data.drop(["loan_id"],axis=1)
print (train_mod.shape)
train_mod.head(3)
```

```python
## Label Encoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
le4=LabelEncoder()
train_mod["source"]=le.fit_transform(train_mod["source"])
train_mod["loan_purpose"]=le1.fit_transform(train_mod["loan_purpose"])
train_mod["financial_institution"]=le2.fit_transform(train_mod["financial_institution"])
train_mod["origination_date"]=le3.fit_transform(train_mod["origination_date"])
train_mod["first_payment_date"]=le4.fit_transform(train_mod["first_payment_date"])
train_mod.head(2)
```

```python
## Splitting Label data

X_sc=train_mod.iloc[:,train_mod.columns!='m13']
print (X_sc.shape)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X_sc)
print (type(X))
new_mod=pd.DataFrame(data=X,columns=X_sc.columns)
print (new_mod.shape)
new_mod.head()
Y=train_mod["m13"]
Y.shape
```

```python
new_mod.head()
```

```python
## Splitting Test Train data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new_mod,Y,test_size=0.25)
y_train.shape
```

```python
#Using SMOTE for Balancing Data

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
x_train_res, y_train_res = sm.fit_sample(x_train, y_train.ravel())
print (sum(y_train_res==0))
print (sum(y_train_res==1))
sum(y_train==1)
```

```python
# Only for SMORT splitting Test Train data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_train_res,y_train_res,test_size=0.25)
print (x_train.shape)
y_train.shape

#x_train=x_train_res.copy()
#y_train=y_train_res.copy()
y_train_res.shape
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
from xgboost import XGBClassifier
xg=XGBClassifier()
xg.fit(x_train,y_train)
pred_xg=xg.predict(x_test)
print (accuracy_score(y_test,pred_xg))
print (confusion_matrix(y_test,pred_xg))
f1_score(y_test,pred_xg)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
pred=dt.predict(x_test)
print (accuracy_score(y_test,pred))
print (confusion_matrix(y_test,pred))
f1_score(y_test,pred)
```

```python
##Logistic Regression
from sklearn.metrics import confusion_matrix,f1_score,accuracy_score
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(x_train, y_train)
y_pred = clf.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
acc_lr = accuracy_score(y_test, y_pred)
print (acc_lr)
print (confusion_matrix(y_test,y_pred))
print (f1_score(y_test, y_pred,average=None))
f1_score(y_test,y_pred)
```

```python
from sklearn.neural_network import MLPClassifier
ml = MLPClassifier(hidden_layer_sizes=(100, 100, 100,))
ml.fit(x_train, y_train)
print("Score: ", ml.score(x_test, y_test))
y_pred_ml= ml.predict(x_test)
#y_pred_proba = ml.predict_proba(x_test,y_pred_ml)
print("F1 score is: {}".format(f1_score(y_test, y_pred_ml)))
#print("AUC Score is: {}".format(roc_auc_score(y_test, y_pred_proba)))
```
